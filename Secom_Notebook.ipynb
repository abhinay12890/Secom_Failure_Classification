```python
from sklearn.feature_selection import SelectKBest,f_classif
```

```python
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import average_precision_score
```

```python
mlflow.set_tracking_uri("file:./mlruns")
```

```python
from sklearn.model_selection import StratifiedKFold, train_test_split

skf=StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,test_size=0.2,random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
import numpy as np
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
top_features=[50,75,100,125,150,175,200]

mlflow.set_experiment("Secom_Classifier")

results={}

for t in top_features:
    with mlflow.start_run(run_name=f"features_{t}"):
        pr_score_list=[]
        threshold_list=[]
        f1_score_list=[]
        for train_idx,val_idx in skf.split(x_train,y_train):
            x_train_split,x_val_split=x_train.iloc[train_idx],x_train.iloc[val_idx]
            y_train_split,y_val_split=y_train.iloc[train_idx],y_train.iloc[val_idx]
            selector=SelectKBest(score_func=f_classif,k=t)
            x_train_selected=selector.fit_transform(x_train_split,y_train_split)
            x_val_selected=selector.transform(x_val_split)

            model=RandomForestClassifier(class_weight="balanced",random_state=42)
            model.fit(x_train_selected,y_train_split)
            y_pred_proba=model.predict_proba(x_val_selected)[:,1]
            pr_score=average_precision_score(y_val_split,y_pred_proba)
            pr_score_list.append(pr_score)
            precision,recall,threshold=precision_recall_curve(y_val_split,y_pred_proba)
            f1_score=(2*precision*recall)/(precision+recall+1e-8)
            best_threshold=threshold[np.argmax(f1_score[:-1])]
            threshold_list.append(best_threshold)
            f1_score_list.append(f1_score[np.argmax(f1_score[:-1])])
        mean_pr_score=np.mean(pr_score_list)
        mean_threshold=np.mean(threshold_list)
        mean_f1_score=np.mean(f1_score_list)
        print(f"Average PR score for {t} features: ",mean_pr_score)
        mlflow.log_param("k_features",t)
        mlflow.log_metric("mean_pr_auc",mean_pr_score)
        results[t]={"mean_pr_score":mean_pr_score,"mean_threshold":mean_threshold,"mean_f1_score":mean_f1_score}
mlflow.end_run()
    
```

```python
results
```

```python
for i,r in results.items():
    print(i,r)
```

```python
best_k=max(results,key=lambda x:results[x]["mean_pr_score"])
print(best_k)
print(results[best_k])
best_threshold=results[best_k]["mean_threshold"]
```

```python
selected_100=SelectKBest(score_func=f_classif,k=best_k)
x_train_selected_100=selected_100.fit_transform(x_train,y_train)
x_test_selected_100=selected_100.transform(x_test)
```

```python
selected_features=selected_100.get_feature_names_out()
selected_features
```

```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
y_train.value_counts()
```

```python
neg=1170
pos=83
```

```python
models_dict={"Random Forest":RandomForestClassifier(class_weight="balanced", random_state=42),
             "XGBoost":XGBClassifier(scale_pos_weight=neg/pos, random_state=42,eval_metric="logloss"),
             "LightGBM":LGBMClassifier(class_weight="balanced",random_state=42),
             "Logistic":Pipeline([("scalar",StandardScaler()),("logistic",LogisticRegression(class_weight="balanced", max_iter=1000))])}
```

```python
model_results={}
```

```python
for n,model in models_dict.items():
    with mlflow.start_run(run_name=f"model_{n}_{best_k}"):
    
        model.fit(x_train_selected_100,y_train)
        y_pred_proba=model.predict_proba(x_test_selected_100)[:,1]
        pr_score=average_precision_score(y_test,y_pred_proba)
        mlflow.log_metric("Average PR Score",pr_score)
        mlflow.sklearn.log_model(model,n)
        model_results[n]=pr_score

mlflow.end_run()
```

```python
model_results
```

```python
best_model=max(model_results,key=lambda x:model_results[x])

print(f"Best Model: {best_model}; PR-AUC: {model_results[best_model]}")
```

```python
from sklearn.metrics import roc_auc_score
```

```python
param_grid = {
    "n_estimators": [100, 200, 300],
    "max_depth": [5, 10, None],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2]
}

import itertools

param_results={}

for n_e,m_d,min_split,min_leaf in list(itertools.product(param_grid["n_estimators"],param_grid["max_depth"],param_grid["min_samples_split"],param_grid["min_samples_leaf"])):
    with mlflow.start_run(run_name=f"rf_{n_e}_{m_d}_{min_split}_{min_leaf}") as run:
        model=RandomForestClassifier(n_estimators=n_e,max_depth=m_d,min_samples_leaf=min_leaf,min_samples_split=min_split,random_state=42,class_weight="balanced")
        model.fit(x_train_selected_100,y_train)
        y_proba=model.predict_proba(x_test_selected_100)[:,1]
        score=average_precision_score(y_test,y_proba)
        ro=roc_auc_score(y_test,y_proba)
        mlflow.log_metric("pr_score",score)
        mlflow.log_metric("roc_score",ro)
        mlflow.log_param("n_estimators",n_e)
        mlflow.log_param("max_depth",m_d)
        mlflow.log_param("min_split",min_split)
        mlflow.log_param("min_leaf",min_leaf)
        mlflow.sklearn.log_model(model,"model")

        param_results[f"{n_e}, {m_d}, {min_split}, {min_leaf}"]={"score":score,"run_id":run.info.run_id}
```

```python
runs=mlflow.search_runs()
runs
```

```python
runs=runs.sort_values("metrics.pr_score",ascending=False)
runs
```

```python
from mlflow.sklearn import load_model

final_model=load_model("models:/Secom_RandomForest/1")

from sklearn.metrics import classification_report,precision_recall_curve
y_final_prob=final_model.predict_proba(x_test_selected_100)[:,1]
y_final_untuned=final_model.predict(x_test_selected_100)

import numpy as np

precision, recall, thresholds = precision_recall_curve(y_test, y_final_prob)

f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)

best_idx = np.argmax(f1_scores[:-1])
best_threshold = thresholds[best_idx]
best_threshold
```

```python
y_final_prob.min(),y_final_prob.max(),y_final_prob.mean()
```

```python
final_model.classes_
```

```python
precision=precision[:-1].tolist()
recall=recall[:-1].tolist()
f1_scores=f1_scores[:-1].tolist()
thresholds=thresholds.tolist()

print(len(thresholds), (np.argmax(f1_scores)))
```

```python
plt.title("Threshold tuning")
plt.plot(thresholds,precision,label='Precision')
plt.plot(thresholds,recall,label="Recall")
plt.plot(thresholds,f1_scores,label="F1 Score")
plt.plot([0.5,0.5],[0,1],linestyle="--",label="default")
plt.plot([best_threshold,best_threshold],[0,1],linestyle="--",label="Optimized")
plt.xlabel("Threshold")
plt.ylabel("Scores")
plt.legend()
plt.show()
```

```python
print("***********************************************")
print("Before Tuning")
print(classification_report(y_test,y_final_untuned))
print("-------------------------------------------")
print("After Tuning")
y_final_tuned=(y_final_prob>=best_threshold).astype(int)
print(classification_report(y_test,y_final_tuned))
print("*******************************************")
```
